```python
import pandas as pd
```

```python
twitter_data = pd.read_csv("twitter_racism_parsed_dataset.csv", quotechar='"').drop(columns = ['id', 'index', "Annotation"])
twitter_data.columns = ['text', 'label']
twitter_data.head(25)
```

```python
twitter_data['longitud'] = twitter_data['text'].str.len()
```

```python
twitter_data = twitter_data[twitter_data['longitud']>60]
```

```python
# Seleccionar 200 filas aleatorias
muestra_aleatoria = twitter_data.sample(n=200, random_state=42)

# Guardar los resultados en un csv
muestra_aleatoria[['text']].to_csv('muestra_racism_dataset.csv',index=False)

# Imprimir los resultados
print("Muestra de 200 filas aleatorias:")
print(muestra_aleatoria)
```

```python
# Seleccionar 25 filas aleatorias
muestra_aleatoria_25 = twitter_data.sample(n=25, random_state=42)

# Guardar los resultados en un csv
muestra_aleatoria_25[['text']].to_csv('muestra_aleatoria_25.csv',index=False)

# Imprimir los resultados
print("Muestra de 25 filas aleatorias:")
print(muestra_aleatoria_25)
```
